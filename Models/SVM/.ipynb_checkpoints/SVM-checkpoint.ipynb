```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python

# Set the working directory
os.chdir(r"C:\\Users\\Vishva\\Desktop\\Preprocessing\\Model")

# Verify the current working directory
print("Current Working Directory:", os.getcwd())
```

```python
data = pd.read_csv('preprocessedDataset(with 11 columns)_standarized.csv')
```

```python
data.isnull()
```

```python
# Drop all columns that contain any null values
data = data.dropna(axis=0)
```

```python
Y = data['Target']
X = data.drop(columns='Target', axis=1)
```

```python
#accuracy score 
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")
```

```python
X_train, X_test, Y_train, Y_test  = train_test_split(X,Y, test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy score of the training data", training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy score of the test data", test_data_accuracy)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
# Apply SMOTE to oversample the minority class in the training set
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, Y_train)

# Check the class distribution before and after SMOTE
print(f"Original class distribution: {Y_train.value_counts()}")
print(f"Resampled class distribution: {y_train_smote.value_counts()}")
```

```python
# Create and train an SVM model
svm_model = SVC(random_state=42)
svm_model.fit(X_train_smote, y_train_smote)

# Make predictions on the test set
y_pred = svm_model.predict(X_test)

# Evaluate the model performance
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# Assuming X and Y are your features and labels
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

# SVM classifier with class_weight balanced
classifier = SVC(class_weight='balanced', kernel='linear')  # You can also use 'rbf' or 'poly' for kernels

# Training the model
classifier.fit(X_train, Y_train)

# Predicting on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

# Predicting on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

# Output accuracies
print(f"Training Data Accuracy: {training_data_accuracy}")
print(f"Test Data Accuracy: {test_data_accuracy}")

# Classification report to check class-wise performance
print(classification_report(Y_test, X_test_prediction))
```
